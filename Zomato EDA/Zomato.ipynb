# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
import warnings
warnings.filterwarnings('ignore')
```

# Reading CSV

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'phone', 'address', 'dish_liked', 'reviews_list', 'menu_item'], axis = 1 )
df.head()
```

```python
df.info()
```

# Droping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace= True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace= True)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.rename(columns= {"approx_cost(for two people)" : "Cost2plates", "listed_in(type)" : "Type"}, inplace= True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

# Listed_in()city and location are both present; let's keep just one of them.

```python
df = df.drop(["listed_in(city)"], axis=1)
```

```python
df.head()
```

```python
df["Cost2plates"].unique()
```

# Removing [comma , ] from cost2plates Column

```python
def handlecomma (value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
df["Cost2plates"] = df['Cost2plates'].apply(handlecomma)
df["Cost2plates"].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending= False)
rest_types
```

```python
rest_type_less_then_1000 = rest_types[rest_types < 1000]
rest_type_less_then_1000
```

```python
def handle_rest_type(value):
    if (value in rest_type_less_then_1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts(ascending= False)

location_lessthan_300 = location[location < 300]

def handle_location(value):
    if (value in location_lessthan_300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

# Cleaning cuisines columns

```python
cuisines = df['cuisines'].value_counts(ascending= False)

cuisines_lessthan_100 = cuisines[cuisines < 100]
cuisines_lessthan_100

def handle_cuisines (value):
    if (value in cuisines_lessthan_100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
Type = df['Type'].value_counts(ascending= False)

Type_lessthan_250 = Type[Type < 250]

def handle_Type (value):
    if (value in Type_lessthan_250):
        return 'other'
    else:
        return value

df['Type'] = df['Type'].apply(handle_Type)
df['Type'].value_counts()
```
